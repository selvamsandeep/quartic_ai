```python
train_sum = pd.DataFrame()
train_cols = train.columns.tolist()
train_sum['column_name'] = train.columns
train_sum['num_rows_none'] = [train[c].isnull().sum() for c in train_cols]
train_sum['percent_of_nones'] = (train_sum.num_rows_none / train.shape[0]).round(4)
train_sum['unique_val_cnt'] = [train[c].nunique() for c in train_cols]
train_sum
```
